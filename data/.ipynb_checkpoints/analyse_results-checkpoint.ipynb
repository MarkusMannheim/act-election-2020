```python
import pandas as pd

data2020 = pd.read_csv("booths2020.csv")
data2016 = pd.read_csv("booths2016.csv")

parties2020 = data2020["party"].unique()
parties2016 = data2016["party"].unique()
booths2020 = data2020["booth"].unique()
booths2016 = data2016["booth"].unique()
booth_results2020 = pd.DataFrame(columns=parties2020, index=booths2020)
booth_results2016 = pd.DataFrame(columns=parties2016, index=booths2016)

for booth in booths2020:
    results = []
    for party in parties2020:
        results.append(data2020[data2020.booth.eq(booth) & data2020.party.eq(party)].votes.sum())
    booth_results2020.loc[booth] = results

booth_results2020["formal votes"] = booth_results2020.sum(axis=1)


for booth in booths2016:
    results = []
    for party in parties2016:
        results.append(data2016[data2016.booth.eq(booth) & data2016.party.eq(party)].votes.sum())
    booth_results2016.loc[booth] = results

booth_results2016["formal votes"] = booth_results2016.sum(axis=1)

for index, row in booth_results2020.iterrows():
    for party in booth_results2020.columns[:-1]:
        booth_results2020.loc[index, party] = row[party] / row["formal votes"]

for index, row in booth_results2016.iterrows():
    for party in booth_results2016.columns[:-1]:
        booth_results2016.loc[index, party] = row[party] / row["formal votes"]
```

```python
for column in booth_results2020.columns[:-1]:
    booth_results2020[column] = booth_results2020[column].astype("float")    
for column in booth_results2016.columns[:-1]:
    booth_results2016[column] = booth_results2016[column].astype("float")
```

```python
booth_details = pd.read_csv("booth_details2020.csv", index_col="Name")
booth_details["lng"] = booth_details["Coordinates"].apply(lambda x: x.split(",")[-1].strip()[5:]).astype("float")
booth_details["lat"] = booth_details["Coordinates"].apply(lambda x: x.split(",")[0].strip()[5:]).astype("float")

locations = []
for booth in booth_results2020.index:
    if booth in booth_details.index:
        locations.append([booth_details.loc[booth, "lng"], booth_details.loc[booth, "lat"]])
    else:
        locations.append("NaN")
booth_results2020["location"] = locations
booth_results2020
```

```python
booth_results2020["other"] = 1 - (booth_results2020["ACT Labor"] + booth_results2020["The ACT Greens"] + booth_results2020["Canberra Liberals"])
results2020 = booth_results2020[["ACT Labor", "Canberra Liberals", "The ACT Greens", "other", "formal votes", "location"]]
results2020.columns = ["labor", "liberals", "greens", "other", "votes", "location"]
results2020.to_csv("results2020.csv", index_label="booth")
results2020
```
