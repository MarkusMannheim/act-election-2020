```python
import pandas as pd

data = pd.read_csv("candidate_votes.csv")
data[["booth", "party", "votes"]].groupby(["booth", "party"]).sum().to_csv("booth_results.csv")
data = pd.read_csv("booth_results.csv")

parties = data["party"].unique()
booths = data["booth"].unique()
booth_results = pd.DataFrame(columns=parties, index=booths)

for booth in booths:
    results = []
    for party in parties:
        results.append(data[data.booth.eq(booth) & data.party.eq(party)].votes.sum())
    booth_results.loc[booth] = results

booth_results["formal votes"] = booth_results.sum(axis=1)

for booth in booth_results.index:
    if "(Early voting)" in booth:
        booth_results.loc[booth.replace(" (Early voting)", "")] = booth_results.loc[booth.replace(" (Early voting)", "")] + booth_results.loc[booth]
        print(booth)

for index, row in booth_results.iterrows():
    for party in booth_results.columns[:-1]:
        booth_results.loc[index, party] = row[party] / row["formal votes"]

# for index, row in booth_results2016.iterrows():
#     for party in booth_results2016.columns[:-1]:
#         booth_results2016.loc[index, party] = row[party] / row["formal votes"]
        
booth_results
```

```python
for column in booth_results.columns[:-1]:
    booth_results[column] = booth_results[column].astype("float")    
# for column in booth_results2016.columns[:-1]:
#     booth_results2016[column] = booth_results2016[column].astype("float")
```

```python
booth_details = pd.read_csv("booth_details2020.csv", index_col="Name")
booth_details["lng"] = booth_details["Coordinates"].apply(lambda x: x.split(",")[-1].strip()[5:]).astype("float")
booth_details["lat"] = booth_details["Coordinates"].apply(lambda x: x.split(",")[0].strip()[5:]).astype("float")
new_index = []
for booth in booth_details.index:
    if "Gungahlin City" in booth:
        booth = booth.replace(" City", "")
    new_index.append(booth)
booth_details.index = new_index

locations = []
for booth in booth_results.index:
    if booth in booth_details.index:
        locations.append([booth_details.loc[booth, "lng"], booth_details.loc[booth, "lat"]])
    else:
        locations.append("NaN")
booth_results["location"] = locations
booth_results
```

```python
booth_results["other"] = 1 - (booth_results["ACT Labor"] + booth_results["The ACT Greens"] + booth_results["Canberra Liberals"])
results = booth_results[["ACT Labor", "Canberra Liberals", "The ACT Greens", "other", "formal votes", "location"]]
results.columns = ["labor", "liberals", "greens", "other", "votes", "location"]
results.to_csv("results.csv", index_label="booth")
results
```
