```python
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# use a headless browser (saves time)
chrome_options = Options()
# chrome_options.add_argument("--headless")

# set up the browser
driver = webdriver.Chrome(options=chrome_options)

# set up scrape
driver.get("https://www.google.com/maps/d/u/0/viewer?mid=1HYZeweBbBDjphj2Iltdfhki5E8h5_iGS&ll=-35.33072930453165%2C149.12141745000002&z=10")
time.sleep(1)
driver.find_element(By.CSS_SELECTOR, ".uVccjd:last-child").click()
time.sleep(1)
booths = driver.find_elements(By.CLASS_NAME, "pbTTYe-ibnC6b-d6wfac")

# scrape booths
booth_data = pd.DataFrame()
for booth in booths:    
    booth.click()
    datum = {}
    booth_details = driver.find_elements(By.CLASS_NAME, "qqvbed-p83tee")
    for detail in booth_details:
        key = detail.find_element(By.CLASS_NAME, "qqvbed-p83tee-V1ur5d").get_attribute("innerText")
        value = detail.find_element(By.CLASS_NAME, "qqvbed-p83tee-lTBxed").get_attribute("innerText")
        datum[key] = value
    booth_data = booth_data.append(pd.Series(datum), ignore_index=True)
    driver.find_element(By.CLASS_NAME, "qqvbed-a4fUwd-LgbsSe-Bz112c").click()
    
booth_data.set_index("Name", inplace=True)
booth_data
```

```python
booth_data.to_csv("booth_detail2020.csv")
```
